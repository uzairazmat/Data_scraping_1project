```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL for Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Imitate a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# Fetch page content
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    html_data = response.text  # Convert response to text
    tesla_revenue = pd.read_html(html_data, match="Tesla Annual Revenue")[0]

  

    # Rename columns for clarity
    tesla_revenue.columns = ["Date", "Revenue"]
    print(tesla_revenue)
else:
    print(f"Error {response.status_code}: Unable to access the page")
```

# Question 5 - Tesla Stock and Revenue Dashboard - 2 Points

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Get Tesla Stock Data
tesla = yf.Ticker("TSLA")
tesla_stock = tesla.history(period="5y")  # Last 5 years

# Step 2: Get Tesla Revenue Data (Replace with actual source)
# Assuming you have a DataFrame with 'Date' and 'Revenue'
tesla_revenue = pd.DataFrame({
    "Date": pd.date_range(start="2020-01-01", periods=10, freq="Q"),  # Example dates (quarterly)
    "Revenue": [6, 8, 10, 12, 15, 18, 20, 22, 25, 28]  # Example revenue in billions
})

# Convert date columns to datetime format
tesla_stock.index = pd.to_datetime(tesla_stock.index)
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

# Step 3: Plot Stock Price and Revenue
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Tesla Stock Price
ax1.plot(tesla_stock.index, tesla_stock["Close"], color="b", label="Stock Price ($)")
ax1.set_xlabel("Date")
ax1.set_ylabel("Stock Price ($)", color="b")
ax1.tick_params(axis="y", labelcolor="b")

# Create second Y-axis for Revenue
ax2 = ax1.twinx()
ax2.plot(tesla_revenue["Date"], tesla_revenue["Revenue"], color="g", marker="o", linestyle="--", label="Revenue (Billion $)")
ax2.set_ylabel("Revenue (Billion $)", color="g")
ax2.tick_params(axis="y", labelcolor="g")

# Titles & Legend
plt.title("Tesla Stock Price vs. Revenue Over Time")
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))
plt.show()
```
