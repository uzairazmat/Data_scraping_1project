```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL for Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Imitate a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# Fetch page content
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    html_data = response.text  # Convert response to text
    tesla_revenue = pd.read_html(html_data, match="Tesla Annual Revenue")[0]

  

    # Rename columns for clarity
    tesla_revenue.columns = ["Date", "Revenue"]
    print(tesla_revenue)
else:
    print(f"Error {response.status_code}: Unable to access the page")
```
